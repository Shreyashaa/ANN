```python
x_input = [0.4, 0.6  , 0.2]
w_weights = [0.8, 0.3, 0.4]
threshold = 0.3
```

```python
perceptron algo:
```

```python
def step (weighted_sum):
  if weighted_sum > threshold :
    return 1
  else: 
    return 0
```

```python
def perceptron():
  weighted_sum = 0 
  for x,w in zip(x_input,w_weights):
    weighted_sum += x*w
    print(weighted_sum)
  return step(weighted_sum)
```

```python
output = perceptron()
print('output:' + str(output))
```
